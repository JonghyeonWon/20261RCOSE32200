```python
data2 ='https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/main/data/6.csv'
```

```python
df = pd.read_csv(data2)
df
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['st'] = pd.to_datetime(df['st'])
```

```python
df['sample1'] = pd.to_datetime(df['sample1'], format = '%Y-%m-%d_%H:%M:%S')
```

```python
df['sample2'] = pd.to_datetime(df['sample2'], format = '%Y-%m-%d %H시')
```

```python
df.info()
```

```python
# 사용 가능한 .dt 메소드
df['dates'].dt.year         # 연도 추출
df['dates'].dt.month        # 월 추출
df['dates'].dt.day          # 일 추출
df['dates'].dt.hour         # 시간 추출
df['dates'].dt.minute       # 분 추출
df['dates'].dt.second       # 초 추출
df['dates'].dt.weekday      # 요일(0=월요일, 6=일요일) 추출
df['dates'].dt.dayofweek    # 요일(0=월요일, 6=일요일) 추출 (weekday와 동일)
df['dates'].dt.dayofyear    # 연도의 몇 번째 날인지 추출
df['dates'].dt.quarter      # 분기 추출
df['dates'].dt.strftime('%Y-%m-%d')  # 포맷에 맞춰 문자열로 변환
df['dates'].dt.to_period('M')  # 주기(period)로 변환 (예: 월 단위)
```
