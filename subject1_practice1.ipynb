```python
# 2회 데이터
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e2_p1_1.csv')
df.head(2)
```

```python
head_index = df['CRIM'].sort_values(ascending=False).head(10).index
min_val = df['CRIM'].sort_values(ascending=False).head(10).min()
df.loc[head_index, 'CRIM'] = min_val
df.loc[df['AGE'] >= 80]['CRIM'].mean()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e2_p1_1.csv')
df_before = df['RM'].dropna().std()
df['RM'] = df['RM'].fillna(df['RM'].median())
df_after = df['RM'].std()
round(abs(df_before - df_after),3)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e2_p1_1.csv')
dis_average = df['DIS'].mean()
range_min = dis_average - (1.5 * df['DIS'].std())
range_max = dis_average + (1.5 * df['DIS'].std())

filter_df = df[(df['DIS'] < range_min) | (df['DIS'] > range_max)]
result = filter_df['DIS'].sum()
# result = 0
# for i in list(df['DIS']):
#   if i < range_min or range_max < i:
#     result += i;
print(result)
```

```python

df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p1_2.csv')
df.head()
```

```python
mean_gdp_1990 = df[(df['Year'] == 1990)]['Value'].mean()
mean_gdp_2010 = df[(df['Year'] == 2010)]['Value'].mean()

gdp_1990_df = df[(df['Year'] == 1990)].reset_index(drop=True)
gdp_2010_df = df[(df['Year'] == 2010)].reset_index(drop=True)

filter_1990 = gdp_1990_df[gdp_1990_df['Value'] <= mean_gdp_1990]['Country Code']
filter_2010 = gdp_2010_df[gdp_2010_df['Value'] >= mean_gdp_2010]['Country Code']

len(set(filter_1990) & set(filter_2010))
```
