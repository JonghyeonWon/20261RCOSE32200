```python
remove_col = ['회원ID', '환불금액']
train_preprocessing = train.copy()
train_preprocessing = train_preprocessing.drop(columns = remove_col)
```

```python
# '범주형'에 대해서 인코딩 필요
# pd.get_dummies(train_preprocessing)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(train_preprocessing['주구매상품'])

le2 = LabelEncoder()
le2.fit(train_preprocessing['주구매지점'])
```

```python
le_predict = le.transform(train_preprocessing['주구매상품'])
le_predict2 = le2.transform(train_preprocessing['주구매지점'])
train_preprocessing['주구매상품'] = le_predict
train_preprocessing['주구매지점'] = le_predict2
train_preprocessing
```

```python
train_preprocessing['주구매상품'].unique()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled_df = train_preprocessing[['총구매액','최대구매액','방문일수','방문당구매건수','구매주기']]
sc = StandardScaler()
sc.fit(scaled_df)
```

```python
sc.transform(scaled_df)
```

```python
train_preprocessing.loc[:,['총구매액','최대구매액','방문일수','방문당구매건수','구매주기']] = sc.transform(scaled_df)
train_preprocessing
```

```python
y = train_preprocessing['성별']
x = train_preprocessing.drop(columns = '성별')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_validation, y_train, y_validation = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.ensemble import XGBClassifier
```

```python
# 모델링 기본 코드
rf = RandomForestClassifier(random_state = 1)
# 모델 학습
rf.fit(x_train, y_train)
# 모델 예측
val_predict = rf.predict(x_validation)
```

```python
xgb = XGBClassifier(random_state = 1)
xgb.fit(x_train, y_train)
val_xgb_predict = xgb.predict(x_validation)
```

```python
# 평가
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score
```

```python
accuracy_score(y_validation, val_predict) # 실제 값, 예측 값
```

```python
accuracy_score(y_validation, val_xgb_predict)
```

```python
val_rf_proba = rf.predict_proba(x_validation)
proba = pd.DataFrame(val_rf_proba)
proba['sum'] = proba.sum(axis = 1)
proba['predict'] = val_predict
proba
```

```python
roc_auc_score(y_validation, val_rf_proba[:,1])
```

```python
f1_score(y_validation, val_predict)
```
