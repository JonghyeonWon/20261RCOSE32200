```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_1.csv')
df.head(5)
```

```python
df['출동시간'] = pd.to_datetime(df['출동일자'].astype('str') + \
                 df['출동시각'].astype('str').str.zfill(6))
df['신고시간'] = pd.to_datetime(df['신고일자'].astype('str') + \
                 df['신고시각'].astype('str').str.zfill(6))
df['소요시간'] = df['출동시간'] - df['신고시간']
df['소요시간'] = df['소요시간'].dt.total_seconds()

new_df = df.groupby('소방서명')['소요시간'].mean()
new_df.sort_values()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_2.csv')
df.head(5)
```

```python
local = df[df['학교세부유형'] == '일반중학교']['시도'].value_counts().index[1]
df_filter = df[(df['시도'] == local) & (df['학교세부유형'] == '일반중학교')]
df_filter['target'] = df['일반학급_학생수_계'] / df['교원수_총계_계']
df_filter.sort_values('target', ascending=False)['교원수_총계_계'].values[0]
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3.csv')
df.head(5)
```

```python
q1 = df['var_6'].quantile(0.25)
q3 = df['var_6'].quantile(0.75)
iqr = q3 - q1
lower = q1 - (1.5 * iqr)
upper = q3 + (1.5 * iqr)
df[(df['var_6'] < lower) | (df['var_6'] > upper)]['var_6'].shape[0]
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p1_2.csv')
df.head(5)
```

```python
df['총방문객'] = df['관광'] + df['방문'] + df['유학연수'] + df['취업'] + df['기타']
df['관광비율'] = df['관광'] / df['총방문객']
target1 = df.sort_values('관광비율',ascending = False)['국가'].values[1]
fir = df[df['국가'] == target1]['취업'].values[0]

target2 = df.sort_values('관광',ascending=False)['국가'].values[2]
sec = df[df['국가'] == target2]['방문'].values[0]

fir + sec
```
