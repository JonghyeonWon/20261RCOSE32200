```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/master/krdatacertificate/10/p1.csv')
df.head()
```

```python
df['총수령액'] = df['기본급'] + df['성과급']
group_df = df.groupby(['년도','기업ID','성별'])['총수령액'].mean()
# group_df.unstack()
groupdf = group_df.reset_index()
pivotdf = groupdf.pivot_table(index = ['년도','기업ID'], columns = '성별', values = '총수령액')
pivotdf['차이'] = abs(pivotdf['남성'] - pivotdf['여성'])
pivotdf['차이'].sort_values().index[-1][1]
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/master/krdatacertificate/10/p2.csv')
df.head()
```

```python
df1 = df[df['구분'] == '검거건수'].reset_index(drop=True)
df2 = df[df['구분'] == '발생건수'].reset_index(drop=True)
dfratio = df1.iloc[:,2:] / df2.iloc[:,2:]
# dfratio['연도'] = df1['연도']

# idxmax 활용
idxmax_df = pd.DataFrame()
idxmax_df['maxtype'] = dfratio.idxmax(axis=1).values
idxmax_df['연도'] = df1['연도']

idxmax_df['index'] = idxmax_df['maxtype'].str.replace('범죄유형', '').astype('int') + 1
result = 0
idx = 0
for i in idxmax_df['index'].values:
  result += df1.iloc[idx, i]
  idx += 1
result
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/master/krdatacertificate/10/p3.csv')
df.head()
```

```python
df['평균만족도'] = df['평균만족도'].fillna(df['평균만족도'].mean())
df['근속연수'] = df['근속연수'].fillna(df.groupby(['부서', '성과등급'])['근속연수'].transform('mean').round())
a = df[(df['부서'] == 'YA') & (df['성과등급'] == 'B')]['근속연수'].mean()
b= df[(df['부서'].str.contains('X')) &(df['평균만족도'] >= 3.2)]['교육참가횟수'].mean()
round(a+b, 3)
```
