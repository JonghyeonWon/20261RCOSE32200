```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/data/2_test.csv')

train
```

```python
train_pre = train.copy()
test_pre = test.copy()
```

```python
train_pre['Mode_of_Shipment'] = train_pre['Mode_of_Shipment'].fillna(train_pre['Mode_of_Shipment'].value_counts().index[0])
train_pre['Cost_of_the_Product'] = train_pre['Cost_of_the_Product'].fillna(train_pre['Cost_of_the_Product'].mean())
train_pre.isnull().sum()
```

```python
train_pre['Customer_care_calls'] = train_pre['Customer_care_calls'].str.replace('$', '').astype('int')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# simple
x = train_pre.drop(columns = ['Reached.on.Time_Y.N', 'ID'])
y = train_pre['Reached.on.Time_Y.N']

x_dummies = pd.get_dummies(x)

x_train, x_val, y_train, y_val = train_test_split(x_dummies, y, test_size = 0.3, random_state = 1)

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf_predict = rf.predict(x_val)

f1_score(y_val, rf_predict)

test_pre = test.copy()
test_pre = test_pre.drop(columns = ['ID'])
test_pre['Customer_care_calls'] = test_pre['Customer_care_calls'].str.replace('$', '').astype('int')

test_dum = pd.get_dummies(test_pre)
rf_test_predict = rf.predict(test_dum)

pd.DataFrame(rf_test_predict, columns = ['pred'])
```

```python
# 라벨 인코딩
train_pre = train.copy()
train_pre['Mode_of_Shipment'] = train_pre['Mode_of_Shipment'].fillna(train_pre['Mode_of_Shipment'].value_counts().index[0])
train_pre['Cost_of_the_Product'] = train_pre['Cost_of_the_Product'].fillna(train_pre['Cost_of_the_Product'].mean())
train_pre['Customer_care_calls'] = train_pre['Customer_care_calls'].str.replace('$', '').astype('int')

le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
le4 = LabelEncoder()

train_pre['Warehouse_block'] = le1.fit_transform(train_pre['Warehouse_block'])
train_pre['Mode_of_Shipment'] = le2.fit_transform(train_pre['Mode_of_Shipment'])
train_pre['Product_importance'] = le3.fit_transform(train_pre['Product_importance'])
train_pre['Gender'] = le4.fit_transform(train_pre['Gender'])

x = train_pre.drop(columns = ['Reached.on.Time_Y.N', 'ID'])
y = train_pre['Reached.on.Time_Y.N']

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.33, random_state = 1)

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf_predict = rf.predict(x_val)

f1_score(y_val, rf_predict)
```

```python
# test 전처리
test_pre = test.copy()
test_pre = test_pre.drop(columns = ['ID'])
test_pre['Customer_care_calls'] = test_pre['Customer_care_calls'].str.replace('$', '').astype('int')

test_pre['Warehouse_block'] = le1.transform(test_pre['Warehouse_block'])
test_pre['Mode_of_Shipment'] = le2.transform(test_pre['Mode_of_Shipment'])
test_pre['Product_importance'] = le3.transform(test_pre['Product_importance'])
test_pre['Gender'] = le4.transform(test_pre['Gender'])

rf_test_predict = rf.predict(test_pre)
result = pd.DataFrame(rf_test_predict, columns = ['pred'])
# result.to_csv('result.csv', index = False)
```
