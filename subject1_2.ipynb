```python
data = 'https://raw.githubusercontent.com/Datamanim/datarepo/refs/heads/main/data/3.csv'
df = pd.read_csv(data)
df
```

```python
# S값이 x, Age가 20이상인 경우
def age_preprocessing(x):
  age = x['Age']
  s = x['S']
  if s =='X' and age >= 20:
    return 'qqq'
  else:
    return 'rrr'

df['age_period_apply'] = df[['Age', 'S']].apply(age_preprocessing, axis = 1)
df
```
